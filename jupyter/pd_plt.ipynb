# Pandas и Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Создаём наборы рандомных данных
ar1 = np.random.randn(50)
ar2 = np.random.randn(50)
ar1
```

```python
ar1.mean(), ar1.min(), ar1.max()
```

```python
plt.plot(ar1, label='Цифры1', c='green') # Обычный график - линия, проходящая через все точки. По Y - значение, по X - номер точки
plt.plot(ar2, label='Цифры2', color='black', linestyle='--')

plt.legend() # Включаем подпись в углу. Угол выбирается автоматически
plt.grid() # Включаем сетку
```

```python
fig, axes = plt.subplots(ncols=1, nrows=1, figsize=(20, 10)) # Создаём сложный график. fig-область, на которой рисуем; axes - оси графика (список осей если графиков много)
ar1_norm = ar1/sum(ar1)
N, bins, _ = axes.hist(ar1_norm, bins=20, label='Цифры1', color='green', rwidth=0.9) # Столбчатая диаграмма (гистограмма). N - высоты столбиков, bins - положения их серединок

axes.legend(fontsize=20) # У всех текстовых полей можно указать размер шрифта
axes.grid()

plt.show() # Выводим картинку на экран. Альтернативно можно использовать plt.savefig() - сохранить как файл 
```

```python
s1 = pd.DataFrame(ar1) # Создаём DataFrame. Создаётся из любого массиво- или словаре-подобного объекта
s1.head()
```

```python
users = [
    {
        'name':'Иван',
        'age':23,
        'is_male': True,
        'dota_hours':5000 
    },
    {
        'name':'Денис',
        'age':18,
        'is_male': True,
        'dota_hours':4000 
    },
    {
        'name':'Ксения',
        'age':17,
        'is_male': False,
        'dota_hours':5 
    },
    {
        'name':'Арсений',
        'age':34,
        'is_male': True,
        'dota_hours':1254 
    },
    {
        'name':'Екатерина',
        'age':25,
        'is_male': False,
    },
]
```

```python
users_df = pd.DataFrame(users)
users_df
```

```python
users_df.describe(include='all') # Выводим статистику по датафрейму
```

```python
users_df.plot(subplots=True, kind='hist', density=True) # Можно сразу представить в виде графиков несколько колонок
```

```python
users_df['age'].plot(grid=True)
```
